```python
def detect_family(col_name):
    c = col_name.lower()
    if 'flex' in c:
        return 'Flex'
    if any(tag in c for tag in ['ax', 'ay', 'az']):
        return 'Accel'
    if any(tag in c for tag in ['yaw', 'pitch', 'roll']):
        return 'Orientation'
    return 'Other'

families = ['Flex', 'Accel', 'Orientation']
family_titles = {
    'Flex': 'Flex Sensors',
    'Accel': 'Accelerometer',
    'Orientation': 'Orientation',
}

fig = make_subplots(
    rows=3, cols=1, shared_xaxes=True, vertical_spacing=0.06,
    subplot_titles=[family_titles[f] for f in families]
)

row_map = {'Flex': 1, 'Accel': 2, 'Orientation': 3}

for col in selected:
    family = detect_family(col)
    if family not in row_map: continue
    
    fig.add_trace(
        go.Scatter(
            x=df[X_COL], y=df[col], mode='lines', name=col,
            hovertemplate="<b>%{fullData.name}</b><br>" + X_COL + "=%{x}<br>Value=%{y}<extra></extra>"
        ),
        row=row_map[family], col=1
    )

fig.update_layout(height=1000, title_text="Hand Tracking Analytics", hovermode='x unified')
fig.show()
```
